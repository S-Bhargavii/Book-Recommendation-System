## Popularity based recommender system 
Display the top 50 books with the highest average rating and atleast 250 ratings ( for it to be authentic)

```python
ratings_with_name = ratings.merge(books,on="ISBN")
```

```python
ratings_with_name
```

```python
num_ratings = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
# reset_index converts into a dataframe 
```

```python
num_ratings.rename(columns={'Book-Rating':'num_ratings'},inplace= True)
```

```python
num_ratings
```

```python
avg_ratings = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
```

```python
avg_ratings.rename(columns={'Book-Rating':'Avg-Rating'},inplace=True)
```

```python
avg_ratings
```

```python
book_ratings = avg_ratings.merge(num_ratings,on='Book-Title')
```

```python
book_ratings.head()
```

```python
book_ratings = book_ratings[book_ratings['num_ratings']>250]
```

```python
book_ratings = book_ratings.sort_values(by = 'Avg-Rating',ascending=False)
```

```python
book_ratings.head(50)
# gives the top 50 highest rated books
```

```python
book_ratings=book_ratings.merge(books,on='Book-Title').drop_duplicates('Book-Title')
```

```python
book_ratings.head(5)
```

```python
popular_books = book_ratings.head(50)
popular_books['Avg-Rating'] = popular_books['Avg-Rating'].round(2)
```

```python
popular_books
```

```python
popular_books = popular_books[['Book-Title','Book-Author','Image-URL-L','Avg-Rating','num_ratings']]
popular_books
```

## Collaborative Filtering 

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
usersp = x[x].index
# consists of the users who have given more than 200 reviews ( they are more credible )
```

```python
usersp
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(usersp)]
```

```python
filtered_rating.head()
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
# filtering out books that have recieved atleast 50 ratings.
```

```python
famous_books = y[y].index 
```

```python
filtered_rating = ratings_with_name[ratings_with_name['Book-Title'].isin(famous_books)]
filtered_rating
```

```python
pivot_table = filtered_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pivot_table
```

```python
pivot_table.fillna(0,inplace=True)
pivot_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score =cosine_similarity(pivot_table)
```

```python
similarity_score 
```

```python
def recommend(book_name):
    distances = similarity_score[np.where(pivot_table.index==book_name)[0][0]]
    similar_items = sorted(list(enumerate(distances)),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        temp_df = books[books['Book-Title']==pivot_table.index[i[0]]]
        item.extend(temp_df.drop_duplicates('Book-Title')['Book-Title'])
        item.extend(temp_df.drop_duplicates('Book-Title')['Book-Author'])
        item.extend(temp_df.drop_duplicates('Book-Title')['Image-URL-L'])
        data.append(item)
    return data
# np.where gives a tuple of the array that contains the indexes of the elements 
# that satisfy the conditions that are written in the parenthesis
#  so the first 0 is to access the array ( the first elemnt of the tuple )
#  the next element is to access the index ( the element of hte array )
# enumerate is a built-in function in Python that is used to iterate over an iterable while 
# keeping track of the index of the current item. It returns a tuple containing the index and the corresponding element.
# it will return an enumerate object --> you should convert into list to view
# books similarity scores
```

```python
recommend('Message in a Bottle')
```
